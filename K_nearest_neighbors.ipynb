```python
import matplotlib.pyplot as plt
import random

X = list(range(40))
#print("X: ", X)
y = []
for i in range(len(X)):
    y.append(20 + X[i] + random.random() * 20)
#print("Y: ", y)
fig, axs = plt.subplots(1, 4, figsize=(15, 4))


def KNN_prediction(k, X, y, x, figure):
    distance = []
    for i in range(len(X)):
        distance.append((X[i], y[i], (x - X[i]) ** 2))
    sorted_list = sort(distance)
    #print(sorted_list)
    x_neigborhood, y_neigborhood = [], []
    y_prediction = 0.0
    for i in range(k):
        y_prediction += sorted_list[i][1]
        x_neigborhood.append(sorted_list[i][0])
        y_neigborhood.append(sorted_list[i][1])
    y_prediction /= k

    plot(figure, X, y, 'y.', "samples")
    plot(figure, x_neigborhood, y_neigborhood, 'k.', "neighborhood")
    plt.plot(x, y_prediction, "k+", label='prediction')
    axs[figure].set_xlabel('{:1.0f} '.format(k) + 'neighbors')
    #print("Y Prediction: ", y_prediction)
    return y_prediction


def sort(unsorted_list):
    return sorted(unsorted_list, key=lambda x: x[2])


def plot(fig, X, y, parameters, label):
    axs[fig].plot(X, y, parameters, label=label)
    axs[fig].legend()
    axs[fig].grid()
    return


KNN_prediction(2, X, y, 7.6, 0)
KNN_prediction(4, X, y, 20.5, 1)
KNN_prediction(7, X, y, 36.2, 2)
KNN_prediction(15, X, y, 24.3, 3)
plt.show()
```
