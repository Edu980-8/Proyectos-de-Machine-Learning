```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def funcion(x):
    return x**2

def deri_func(x):
    return 2*x
def tangent_line(x,x0, y0,slope):
    return slope * (x-x0)+y0

def gradient_descent(x_point, learning_rate, num_iterations):
    points=[x_point]
    for _ in range(num_iterations): 
        gradient = deri_func(x_point)
        x_point = x_point- learning_rate*gradient
        points.append(x_point)
    return points
```

```python
x = np.linspace(-10,10,100)
f_x = funcion(x)
```

```python
x_init = -6.5
f_x_init = funcion(x_init)
slope = deri_func(x_init)
tangent = tangent_line(x,x_init,f_x_init,slope)

print(slope)

plt.plot(x,f_x, label= "Funcion")
plt.plot(x_init,f_x_init,'o', label= "Punto Inicial", markersize=12)
plt.plot(x,tangent,'g--', label="Recta Tangente")
plt.plot(points,[funcion(x) for x in points], 'ro-', label="Descenso de gradiente")
plt.plot(points[-1],funcion(points[-1]), 'bo', "ltimo Punto")


plt.title("Gradiente Descendente")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.xlim(-10,10)
plt.ylim(-10,100)
plt.legend()
plt.grid(True)
plt.show

points = gradient_descent(x_init, learning_rate = 0.1, num_iterations = 20)
```
